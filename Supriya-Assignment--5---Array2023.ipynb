```python
#Question 1

Convert 1D Array Into 2D Array

You are given a **0-indexed** 1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with
creating a 2-dimensional (2D) array with  m rows and n columns using **all** the elements from original.

The elements from indices 0 to n - 1 (**inclusive**) of original should form the first row of the constructed 2D array,
the elements from indices n to 2 * n - 1 (**inclusive**) should form the second row of the constructed 2D array, and so on.

Return *an* m x n *2D array constructed according to the above procedure, or an empty 2D array if it is impossible*.

**Example 1:** 
    
**Input:** original = [1,2,3,4], m = 2, n = 2

**Output:** [[1,2],[3,4]]

**Explanation:** The constructed 2D array should contain 2 rows and 2 columns.

The first group of n=2 elements in original, [1,2], becomes the first row in the constructed 2D array.

The second group of n=2 elements in original, [3,4], becomes the second row in the constructed 2D array.

"""To solve this problem, we need to iterate through the original 1D array and form the 2D array by grouping the elements 
   based on the given row and column dimensions (m and n). Here's the algorithm to convert the 1D array into a 2D array:

  1. Check if the total number of elements in the original array is equal to m * n. If not, return an empty 2D array.

  2. Create an empty 2D array called result with m rows and n columns.

  3. Iterate through the original array using a loop variable i from 0 to the length of the original array minus 1.

  4. Calculate the row and column indices for each element using the formula:
     . row_index = i / n (integer division)
     . column_index = i % n
     
  5. Place the element at the corresponding row and column index in the result array.

  6. Return the result array."""

#Here's the implementation of the algorithm in Python:

def convert_to_2d_array(original, m, n):
    if len(original) != m * n:
        return []
    
    result = [[0] * n for _ in range(m)]
    
    for i in range(len(original)):
        row_index = i // n
        column_index = i % n
        result[row_index][column_index] = original[i]
    
    return result

#Using the example given:

original = [1, 2, 3, 4]
m = 2
n = 2
result = convert_to_2d_array(original, m, n)
print(result)

#As expected, the original 1D array [1, 2, 3, 4] is converted into a 2D array [[1, 2], [3, 4]].
```

```python
#<aside>
ðŸ’¡ **Question 2**

You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has
exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

**Example 1:**

</aside>

**Input:** n = 5

**Output:** 2

**Explanation:** Because the 3rd row is incomplete, we return 2.

"""To solve this problem, we need to determine the number of complete rows that can be built using n coins. Since each row 
   i has i coins, we can iterate over the rows starting from 1 and subtract the corresponding number of coins from n until 
   we can no longer form a complete row.

  Here's the algorithm to find the number of complete rows:

  1. Initialize a variable complete_rows to 0 to keep track of the number of complete rows.

  2. Initialize a variable remaining_coins to n to keep track of the remaining coins.

  3. Initialize a variable current_row to 1 to represent the current row number.

  4. While remaining_coins is greater than or equal to current_row:
    . Increment complete_rows by 1.
    . Subtract current_row from remaining_coins.
    . Increment current_row by 1.
    
  5. Return the value of complete_rows."""

#Here's the implementation of the algorithm in Python:

def count_complete_rows(n):
    complete_rows = 0
    remaining_coins = n
    current_row = 1
    
    while remaining_coins >= current_row:
        complete_rows += 1
        remaining_coins -= current_row
        current_row += 1
    
    return complete_rows

#Using the example given:

n = 5
result = count_complete_rows(n)
print(result)

#As expected, with 5 coins, we can build 2 complete rows (1 row with 1 coin and 1 row with 2 coins), and the 3rd row would 
be incomplete.
```

```python
#<aside>
ðŸ’¡ **Question 3**

Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in 
non-decreasing order*.

**Example 1:**

**Input:** nums = [-4,-1,0,3,10]

**Output:** [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].

After sorting, it becomes [0,1,9,16,100].

</aside>

"""To solve this problem, you can follow these steps:

  1. Create a new array to store the squared values.
  2. Iterate over each number in the input array, nums.
  3. Square each number and add it to the new array.
  4. Sort the new array in non-decreasing order.
  5. Return the sorted array."""

#Here's the implementation in Python:

def sortedSquares(nums):
    squares = [num**2 for num in nums]  # Square each number
    squares.sort()  # Sort the squared values
    return squares 

#Now, let's test the function with the provided example:

nums = [-4, -1, 0, 3, 10]
result = sortedSquares(nums)
print(result)

#The function correctly squares each number in the input array and returns a new array with the squared values sorted in
non-decreasing order.
```

```python
#<aside>
ðŸ’¡ **Question 4**

Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. 
Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. 
Therefore, answer[1] = [4,6].

</aside>

"""To solve this problem, you can follow these steps:

  1. Convert both input arrays, nums1 and nums2, to sets to remove duplicates and facilitate membership testing.
  
  2. Initialize two empty lists, not_in_nums2 and not_in_nums1, to store the distinct integers that are not present in 
     nums2 and nums1, respectively.
     
  3. Iterate over each element in nums1. If the element is not present in the set representation of nums2, add it to
     not_in_nums2.
     
  4. Iterate over each element in nums2. If the element is not present in the set representation of nums1, add it to 
     not_in_nums1.
     
  5. Return a list [not_in_nums1, not_in_nums2] as the result."""

#Here's the implementation in Python:

def findDisappearedNumbers(nums1, nums2):
    nums1_set = set(nums1)
    nums2_set = set(nums2)
    not_in_nums2 = [num for num in nums1 if num not in nums2_set]
    not_in_nums1 = [num for num in nums2 if num not in nums1_set]
    return [not_in_nums1, not_in_nums2]

#Now, let's test the function with the provided example:

nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
result = findDisappearedNumbers(nums1, nums2)
print(result)

"""The function correctly identifies the distinct integers that are not present in nums2 for answer[0] and the distinct
integers that are not present in nums1 for answer[1].
"""
```

```python
#<aside>
ðŸ’¡ **Question 5**

Given two integer arrays arr1 and arr2, and the integer d, *return the distance value between the two arrays*.

The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where
|arr1[i]-arr2[j]| <= d.

**Example 1:**

**Input:** arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

**Output:** 2

**Explanation:**

For arr1[0]=4 we have:

|4-10|=6 > d=2

|4-9|=5 > d=2

|4-1|=3 > d=2

|4-8|=4 > d=2

For arr1[1]=5 we have:

|5-10|=5 > d=2

|5-9|=4 > d=2

|5-1|=4 > d=2

|5-8|=3 > d=2

For arr1[2]=8 we have:

**|8-10|=2 <= d=2**

**|8-9|=1 <= d=2**

|8-1|=7 > d=2

**|8-8|=0 <= d=2**

</aside>

"""To solve this problem, we can iterate over each element in arr1 and check if there exists any element in arr2 that
   satisfies the condition |arr1[i] - arr2[j]| <= d. If no such element exists, we increment a counter variable. Finally, 
   we return the value of the counter."""

#Here's the implementation in Python:

def distance_value(arr1, arr2, d):
    count = 0

    for num1 in arr1:
        found = False
        for num2 in arr2:
            if abs(num1 - num2) <= d:
                found = True
                break

        if not found:
            count += 1

    return count

#Let's test the function with the given example:

arr1 = [4, 5, 8]
arr2 = [10, 9, 1, 8]
d = 2

print(distance_value(arr1, arr2, d))

#The output is 2, which matches the expected result.
```

```python
#<aside>
ðŸ’¡ **Question 6**

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears 
**once** or **twice**, return *an array of all the integers that appears **twice***.

You must write an algorithm that runs in O(n) time and uses only constant extra space.

**Example 1:**

**Input:** nums = [4,3,2,7,8,2,3,1]

**Output:**

[2,3]

</aside>

"""To find the integers that appear twice in the given array nums, we can utilize the fact that all the integers in the 
   array are in the range [1, n]. This means we can use the indices of the array to mark the presence of each number.

   Here's an algorithm to solve the problem in O(n) time complexity and using constant extra space:

   1. Initialize an empty list called result to store the numbers that appear twice.
   2. Iterate over each number num in the array nums.
   3. Calculate the index idx as the absolute value of num minus 1.
   4. If nums[idx] is positive, set it to its negation to mark the presence of num.
   5. If nums[idx] is already negative, it means num has already been marked as a duplicate. Append num to the result list.
   6. After iterating over all numbers, return the result list."""

#Here's the implementation of the algorithm in Python:

def findDuplicates(nums):
    result = []

    for num in nums:
        idx = abs(num) - 1
        if nums[idx] > 0:
            nums[idx] = -nums[idx]
        else:
            result.append(abs(num))

    return result

#Let's test the function with the given example:

nums = [4, 3, 2, 7, 8, 2, 3, 1]
print(findDuplicates(nums))

#The output is [2, 3], which matches the expected result. These are the integers that appear twice in the array.
```

```python
#<aside>
ðŸ’¡ **Question 7**

Suppose an array of length n sorted in ascending order is **rotated** between 1 and n times. For example, the array nums
= [0,1,2,4,5,6,7] might become:

- [4,5,6,7,0,1,2] if it was rotated 4 times.
- [0,1,2,4,5,6,7] if it was rotated 7 times.

Notice that **rotating** an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], 
..., a[n-2]].

Given the sorted rotated array nums of **unique** elements, return *the minimum element of this array*.

You must write an algorithm that runs in O(log n) time.

**Example 1:**

**Input:** nums = [3,4,5,1,2]

**Output:** 1

**Explanation:**

The original array was [1,2,3,4,5] rotated 3 times.

</aside>

"""To find the minimum element in a sorted rotated array with unique elements in O(log n) time, we can use a modified binary
   search algorithm.

  The idea is to divide the array into two parts and determine which part contains the minimum element. We can compare the
  middle element with the first and last elements to determine which part of the array is sorted.

  Here's the algorithm to find the minimum element:

  1. Initialize two pointers, left and right, to the start and end indices of the array, respectively.
  
  2. Check if the element at index left is less than the element at index right. If true, it means the array is already sorted,
     and the minimum element is at index left. Return the element.

  3. Perform binary search until left is less than right:
      . Calculate the middle index as mid = (left + right) // 2.
      . Check if the middle element is greater than the element at the right index. If true, it means the minimum element is 
        in the right half of the array. Update left = mid + 1.
      . Otherwise, the minimum element is in the left half of the array. Update right = mid.
   4. The algorithm will converge to a single element, which is the minimum element. Return the element at the left index."""

#Here's the Python implementation of the algorithm:

def findMin(nums):
    left, right = 0, len(nums) - 1

    # Check if the array is already sorted
    if nums[left] < nums[right]:
        return nums[left]

    # Perform binary search
    while left < right:
        mid = (left + right) // 2

        # Check if the minimum element is in the right half
        if nums[mid] > nums[right]:
            left = mid + 1
        # Otherwise, it's in the left half or at the middle index
        else:
            right = mid

    return nums[left]

#Now, let's test the algorithm with the provided example:

nums = [3, 4, 5, 1, 2]
print(findMin(nums))

#The algorithm correctly returns the minimum element as 1.
```

```python
#<aside>
ðŸ’¡ **Question 8**

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element

in original, and then randomly **shuffling** the resulting array.
Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return 
an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

</aside>

"""To solve this problem, we can iterate over each element in the changed array and check if its half value exists in a set. 
   If it does, we remove the half value from the set since it has been matched. After iterating over all elements, if the set
   is empty, it means all elements in changed have corresponding halves in original. We can then return the values in the set
   as the original array.

  Here's the algorithm to solve the problem:

  1. Create an empty set called originalSet.
  2. Iterate over each element num in the changed array:
     . Check if num/2 exists in the originalSet:
        . If it exists, remove num/2 from the originalSet.
        . If it doesn't exist, return an empty array, as the changed array is not a doubled array.
     . Add num to the originalSet.
  3. If the originalSet is empty, return an empty array.
  4. Otherwise, return the elements in the originalSet as the original array."""

#Here's the Python implementation of the algorithm:

def findOriginalArray(changed):
    n = len(changed)
    if n % 2 != 0:  # If the length is odd, it's not a valid doubled array
        return []

    originalSet = set()
    for num in changed:
        if num / 2 in originalSet:
            originalSet.remove(num / 2)
        else:
            return []

        originalSet.add(num)

    if len(originalSet) == 0:
        return []

    return list(originalSet)

#Let's test the algorithm with the provided example:

changed = [1, 3, 4, 2, 6, 8]
print(findOriginalArray(changed))

#The algorithm correctly returns one possible original array, which is [1, 3, 4].
```
